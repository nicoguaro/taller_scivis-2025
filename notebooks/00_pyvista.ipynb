# Ejemplo de PyVista

PyVista es una biblioteca auxiliar para el Kit de Herramientas de Visualización (VTK)
que ofrece un enfoque diferente para la interacción con VTK mediante NumPy y acceso
directo a arreglos.

Para trabajar de forma interactiva en Jupyter con PyVista se requiere usar
[Trame](https://kitware.github.io/trame/) y sus complementos para Jupyter.
Estos se pueden instalar con

```console
pip install "pyvista[jupyter]"
````

`````python
import numpy as np

import pyvista as pv
from pyvista import examples
`````

`````python
pv.set_jupyter_backend('trame')
`````

`````python
# Make data
xrng = np.linspace(-10, 10)
yrng = np.linspace(-10, 10, 20)
x, y = np.meshgrid(xrng, yrng, indexing='ij')
r = np.sqrt(x**2 + y**2)
z = np.sin(r)
`````

`````python
# Create and plot structured grid
grid = pv.StructuredGrid(x, y, z)
grid.plot(smooth_shading=True)
`````
