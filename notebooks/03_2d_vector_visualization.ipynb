# Vector data visualization in two dimensions

Para datos bidimensionales, es común asociar las dimensiones de los datos con la dimensión en pantalla.

Presentamos dos métodos para la visualización vectorial:

- Glifos: Utilizamos símbolos (generalmente flechas) para representar la magnitud y la dirección del vector.
- Líneas de corriente: Presentamos líneas tangentes a los vectores en cada punto.

```python
%matplotlib widget
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use("../img/matplotlib.mplstyle")
```

```python
def ex_data(xmin=-2, xmax=2, ymin=-2, ymax=2, nx=21, ny=21):
    x = np.linspace(xmin, xmax, nx)
    y = np.linspace(ymin, ymax, nx)
    X, Y = np.meshgrid(x, y)
    U = Y
    V = -X - X**2
    return X, Y, U, V
```

```python
X, Y, U, V = ex_data()
```

## Visualización con glifos



```python
plt.figure(figsize=(6, 6))
plt.quiver(X, Y, U, V)
plt.axis("image")
plt.xlabel("$x$")
plt.ylabel("$y$");
```

Observe que en el ejemplo anterior codificamos tanto la magnitud como la dirección en cada símbolo.
También podemos considerar usar las flechas solo para representar la dirección y usar una de las
técnicas de visualización escalar para la magnitud.

```python
mag = np.sqrt(U**2 + V**2)
U_norm = U/(mag + 1e-6) # Queremos evitar la división por cero
V_norm = V/(mag + 1e-6)
plt.figure(figsize=(6, 6))
plt.contourf(X, Y, mag, cmap="summer")
plt.quiver(X, Y, U_norm, V_norm)
plt.axis("image")
plt.xlabel("$x$")
plt.ylabel("$y$");
```

## Streamlines

Por defecto, la magnitud del campo vectorial está representada por la densidad de las líneas.

```python
plt.figure(figsize=(6, 6))
plt.streamplot(X, Y, U, V)
plt.axis("image")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```

También podemos cambiar el ancho de la línea para representar la magnitud (u otra cantidad).

```python
plt.figure(figsize=(6, 6))
plt.streamplot(X, Y, U, V, linewidth=0.5*mag)
plt.axis("image")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```

También podemos cambiar el color para representar la magnitud (u otra cantidad).

```python
plt.figure(figsize=(6, 6))
plt.streamplot(X, Y, U, V, color=mag)
plt.axis("image")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```
