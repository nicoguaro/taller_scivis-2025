# Ejemplo de IPyVolume

IPyVolume nos permite visualizar volúmenes y glifos 3D en cuadernos de Jupyter, con una configuración y
un esfuerzo mínimos.

Para más información se puede consultar la documentación del proyecto: https://ipyvolume.readthedocs.io

```python
import numpy as np
import ipyvolume as ipv
```

```python
x, y, z = np.ogrid[-10:10:20j, -10:10:20j, -10:10:20j]
s = np.sin(x*y*z)/(x*y*z)
```

```python
ipv.quickvolshow(s)
```
